```python
#Extraccion de datos de mi base limpia connlas columnas seleccionadas
#librerias y subida de archivo
import pandas as pd
df_valencia=pd.read_csv("Valencia_España_listings_50cols.csv")
```

```python
#Los registros de 5 “host name” diferentes
unique_host_id = df_valencia["host_id"].unique()
print(unique_host_id[:5])
df_valencia.to_csv("unique_host_id", index=False)
```

```python
# Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
df_valencia['host_since'] = pd.to_datetime(df_valencia['host_since'], format="%d/%m/%Y")
df_filtered = df_valencia[df_valencia["host_since"].dt.year > 2020]
# Guardar a CSV
df_filtered.to_csv("df_filtered.csv", index=False)
# Mostrar resultado
df_filtered
```

```python
#Los registros de los host que responden “a más tardar en 1 día”
df_filtered = df_valencia[df_valencia["host_response_time"] == "within a day"]
df_filtered
#save csv
df_filtered.to_csv("df_filtered_host_response_time.csv", index=False)
```

```python
#Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
df_d = df_valencia[(df_valencia.index % 200 == 0) & (df_valencia.index != 0)].copy()
# Guardar resultado
df_d.to_csv("/content/filtro_d_multiplos_200.csv", index=False)
df_d
```

```python
# Los registros que “a lo mucho cuenten con 4 camas”
df_filtered = df_valencia[df_valencia["beds"] <= 4]
#save csv
df_filtered.to_csv("df_filtered_beds.csv", index=False)
df_filtered
```

```python
#Los registros que “sean superhost”
df_filtered = df_valencia[df_valencia["host_is_superhost"] == "t"]
df_filtered
#save csv
df_filtered.to_csv("df_filtered_superhost.csv", index=False)
df_filtered
```

```python
#Los registros de las columnas “impares”
df_impares = df_valencia.iloc[:, 1::2]
df_impares
#save csv
df_impares.to_csv("df_impares.csv", index=False)
df_impares
```
