```python
import pandas as pd
df= pd.read_csv("Datos_FacturaciÃ³n.csv")
```

```python
#Filtro 1 CVE CLPV
filtro1=df[(df["CVE_CLPV"] >= 1000) & (df["CVE_VEND"] <= 2000)]
#Save csv
filtro1.to_csv("Filtro1.csv")
filtro1
```

```python
#Filtro 2 CVE_VEND(Todas las claves excepto 5 y 4)
filtro2=df[(df["CVE_VEND"] != 5) & (df["CVE_VEND"] != 4)]
#Save csv
filtro2.to_csv("Filtro2.csv")
filtro2
```

```python
#Filtro 3 fechas de 28/02/2022
filtro3=df[(df["FECHA_ENT"] == "2022-02-28 00:00")]
#Save csv
filtro3.to_csv("Filtro3.csv")
filtro3
```

```python
#Filtro4 CAN_TOT (Cantidades menores a 5951.7) oSTATUS (âEâ)
filtro4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
#Save csv
filtro4.to_csv("Filtro4.csv")
filtro4
#
```

```python
#Filtro5Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5=df.iloc[:,[1,7,8,10]]
#Save csv
filtro5.to_csv("Filtro5.csv")
filtro5
```

```python
#Solo las filas de 7001-7099
filtro6=df.iloc[7001:7100,:]
#Save csv
filtro6.to_csv("Filtro6.csv")
filtro6
```

```python
#filtro 7 Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)
filtro7=df[(df["CVE_VEND"] == 1) | (df["CVE_VEND"] == 2)]
filtro7=filtro7.set_index("CVE_VEND")
filtro7
```
